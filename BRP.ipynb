```python
from google.colab import drive
import os
import zipfile

# Google Drive 마운트
drive.mount('/content/drive')

def unzip_from_drive(zip_path, extract_path='/content'):
    """
    Google Drive의 zip 파일을 Colab의 지정된 경로에 압축해제하는 함수

    Args:
        zip_path (str): Drive 내의 zip 파일 경로 (예: '/content/drive/MyDrive/data.zip')
        extract_path (str): 압축을 해제할 경로 (기본값: '/content')
    """
    try:
        # zip 파일이 존재하는지 확인
        if not os.path.exists(zip_path):
            raise FileNotFoundError(f"Zip 파일을 찾을 수 없습니다: {zip_path}")

        # 압축 해제
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)

        print(f"압축 해제 완료: {extract_path}")

        # 압축 해제된 파일/폴더 목록 출력
        print("\n압축 해제된 파일/폴더 목록:")
        for root, dirs, files in os.walk(extract_path):
            level = root.replace(extract_path, '').count(os.sep)
            indent = ' ' * 4 * level
            print(f"{indent}{os.path.basename(root)}/")
            subindent = ' ' * 4 * (level + 1)
            for f in files:
                print(f"{subindent}{f}")

    except Exception as e:
        print(f"오류가 발생했습니다: {str(e)}")

# 사용 예시
zip_path = "/content/drive/MyDrive/아네리카노/face_aligned.zip"
audio_zip_path = '/content/drive/MyDrive/아네리카노/audio.zip'
unzip_from_drive(zip_path)
unzip_from_drive(audio_zip_path)
```
